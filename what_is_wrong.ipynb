```python
import tensorflow as tf
import pandas as pd
```

```python
def sum_log(path):
    runlog = pd.DataFrame(columns=['metric', 'value'])
    
    for e in tf.compat.v1.train.summary_iterator(path):
        for v in e.summary.value:
            r = {'metric': v.tag, 'value':v.simple_value}
            runlog = runlog.append(r, ignore_index=True)

    return runlog
```

```python
path = "./runs/Aug20_13-00-08_DESKTOP-2OM3QF1/events.out.tfevents.1629482409.DESKTOP-2OM3QF1.8064.0"

longer = sum_log(path)

wider = pd.DataFrame({
    "discriminator_loss": longer.loc[longer.metric=='loss_discriminator', :].reset_index().value,
    "generator_loss": longer.loc[longer.metric=='loss_generator', :].reset_index().value,
    "act_mean": longer.loc[longer.metric=='act_mean', :].reset_index().value,
    "act_std": longer.loc[longer.metric=='act_std', :].reset_index().value,
})
```
