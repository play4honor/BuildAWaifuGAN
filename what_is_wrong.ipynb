```python
import tensorflow as tf
import pandas as pd
```

```python
def sum_log(path):
    runlog = pd.DataFrame(columns=['metric', 'value'])
    
    for e in tf.compat.v1.train.summary_iterator(path):
        for v in e.summary.value:
            r = {'metric': v.tag, 'value':v.simple_value}
            runlog = runlog.append(r, ignore_index=True)

    return runlog
```

```python
path = "./runs/Aug19_20-45-46_DESKTOP-2OM3QF1/events.out.tfevents.1629423948.DESKTOP-2OM3QF1.7636.0"

longer = sum_log(path)

wider = pd.DataFrame({
    "discriminator_loss": longer.loc[longer.metric=='loss_discriminator', :].reset_index().value,
    "grad_penalty": longer.loc[longer.metric=='grad_penalty', :].reset_index().value,
    "non_grad_loss": longer.loc[longer.metric=='non_grad_loss', :].reset_index().value,
    "discriminator_activation": longer.loc[longer.metric=='non_grad_loss', :].reset_index().value,
    "generator_loss": longer.loc[longer.metric=='dis_activation', :].reset_index().value,
    ""
})
```
